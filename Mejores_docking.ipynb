```python
# Definir puntajes de corte
my_dick = {"EGFR": -9.29, "MAPK1": -10.15, "SRC": -7.8, "STAT3": -9.41}
```

```python
import pandas as pd

df_bases = pd.read_csv("Resultados_docking/Bases_conjuntas/Duales.csv", index_col=0)
df_inhib_dual = pd.read_csv("Resultados_docking/Inhibidores/Duales.csv", index_col=0)
df_inhib_triples = pd.read_csv("Resultados_docking/Inhibidores/Triples.csv", index_col=0)
```

# Analizar bases conjuntas
<hr>

```python
df_bases["Targets"] = df_bases["Target_1"] + "_" + df_bases["Target_2"]
```

```python
df_bases["Targets"].value_counts()
```

```python
# Obtener aquellos mejores o iguales a EGFR y MAPK1

df_EGFR_MAPK1 = df_bases[df_bases["Targets"] == "EGFR_MAPK1"]
df_EGFR_MAPK1_mejores = df_EGFR_MAPK1[(df_EGFR_MAPK1["Afinidad_1"] <= my_dick["EGFR"]) & (df_EGFR_MAPK1["Afinidad_2"] <= my_dick["MAPK1"])]
```

```python
df_EGFR_MAPK1_mejores
```

```python
# Obtener aquellos mejores o iguales entre EGFR y SRC
# Obtener aquellos mejores o iguales a EGFR y SRC
df_EGFR_SRC = df_bases[df_bases["Targets"] == "EGFR_SRC"]

df_EGFR_SRC_mejores = df_EGFR_SRC[
    (df_EGFR_SRC["Afinidad_1"] <= my_dick["EGFR"]) &
    (df_EGFR_SRC["Afinidad_2"] <= my_dick["SRC"])]
```

```python
df_EGFR_SRC_mejores
```

```python
# Obtener aquellos mejores o iguales a EGFR y STAT3
df_EGFR_STAT3 = df_bases[df_bases["Targets"] == "EGFR_STAT3"]

df_EGFR_STAT3_mejores = df_EGFR_STAT3[
    (df_EGFR_STAT3["Afinidad_1"] <= my_dick["EGFR"]) &
    (df_EGFR_STAT3["Afinidad_2"] <= my_dick["STAT3"])
]
```

```python
df_EGFR_STAT3_mejores
```
