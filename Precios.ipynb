# Cargar datos
<hr>

```python
import pandas as pd
```

```python
df_duales = pd.read_csv("df_mejores/Inhibidores/duales.csv")
df_inhib_triples = pd.read_csv("df_mejores/inhibidores/triples.csv")
df_Bases = pd.read_csv("df_mejores/Bases/EGFR_MAPK1.csv")
```

```python
df_Bases
```

# Cargar referencias
<hr>

```python
IDS_bases = pd.read_csv("IDS/df_bases.csv")
IDS_inhibidores = pd.read_csv("IDS/df_inhibidores.csv")
```

```python
IDS_bases
```

```python
df_Bases.rename(columns={"Molecula": "ID"}, inplace=True)
df_inhib_triples.rename(columns={"Molecula": "ID"}, inplace=True)
df_duales.rename(columns={"Molecula": "ID"}, inplace=True)
```

```python
merge_bases = pd.merge(df_Bases, IDS_bases, on="ID", how="left")
merge_inhib_triples = pd.merge(df_inhib_triples, IDS_inhibidores, on="ID", how="left")
merge_duales = pd.merge(df_duales, IDS_inhibidores, on="ID", how="left")
```

```python
merge_bases
```

# Guardar datos
<hr>

```python
merge_bases.to_excel("Precios/merge_bases.xlsx", index=False)
merge_inhib_triples.to_excel("Precios/merge_inhib_triples.xlsx", index=False)
merge_duales.to_excel("Precios/merge_duales.xlsx", index=False)
```
